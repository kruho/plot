複数の折れ線グラフを重ねて描く

```python
import numpy as np

# x=0~10の範囲を100等分し、sin(x)を生成
xmin = 0
xmax = 10
nb_split = 100
sin = np.sin(np.linspace(xmin, xmax, nb_split))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# 100個の折れ線グラフを重ねる
for i in range(100):
    # sin(x)にガウシアンノイズN(0,1)を乗せる。
    rand = np.random.normal(size=nb_split)
    
    # 線の色は青（b=Blue）、特に工夫せずに重ねて描画。
    plt.plot(sin+rand, c='b')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

for i in range(100):
    rand = np.random.normal(size=nb_split)
    
    # アルファチャンネルを0.05にして透過させる。
    plt.plot(sin+rand, c='b', alpha=0.05)
```
